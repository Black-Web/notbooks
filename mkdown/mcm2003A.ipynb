# MCM2003 A题分析

# Assumptions

* 人和车在下降过程中除重力外不受其他力的作用，即忽略空气阻力，根据这一假设计算的速度值，相比于计及空气阻力时要大些，偏于安全。

* 特技演员在整个过程中始终保持对车的掌控，即人/车在落地后依然保留水平速度。

# Notations

$m$ : 人/车的总质量

$g$ : 重力常量，9.8 $m/s^2$

$F_{m}$ : 人能承受的最大冲击力，受力大于此值时会有危险

$l$ : 直升机螺旋桨距离人/车离开楼面处的水平最远距离

$h$ : 直升机螺旋桨距离人/车的垂直距离

$H$ : 直升机螺旋桨距离地面的高度

$v_0$ : 人/车离开楼面时的初速度

$v_y$ : 接触箱子瞬间的竖直速度

$v_{yR}$ : 接触地面前的瞬时速度

$F_y$ : 箱子对人的平均垂直作用力

$\Delta t_1$ : 人/车与箱子接触后到接触地面前的时间间隔

$\Delta t_2$ : 人/车接触地面后垂直速度从$v_{yR}$降至0的时间间隔

# 分析过程

落地过程可以分为3个阶段讨论

## 阶段1 : 空中只受重力的平抛运动

此过程由方程组（1）描述：

* 根据牛顿运动定律，下降高度为$h$，设下降时间为$t$，则有：
$$1/2*g*t^2 = h$$
* 平抛运动的水平运动距离应大于直升机螺旋桨距离人/车离开楼面处的水平最远距离，即：
$$v_0*t > l$$
* 根据能力守恒，在与箱子接触瞬间，满足：
$$1/2*m{v_y}^2 = mg(H+h)——(4)$$
注意到这里忽略了箱子的高度，实际高度要比$H+h$小些，故根据上式的计算结果偏于安全。

## 阶段2 : 接触到箱子的瞬间到接触到地面之前

由于假定人/车在落地后依然保留水平速度，即对人造成冲击的力主要来自于竖直方向，故这里只讨论竖直方向上的受力情况。

* 竖直速度由$v_y$变为$v_{yR}$的过程，箱子对人/车的竖直作用力满足动量守恒:
$$F_y*\Delta t_1 = m(v_y-v_{yR})——(3)$$

## 阶段3 : 接触地面使得竖直速度骤降至0

这一过程中，为保证人所受冲击力小于$F_m$，应有：
$$F_m * \Delta t_2 >= mv_{yR}——(2)$$

# 分析结果

## 由方程组(1)可以得到：

$$v_0 > \sqrt{{l^2g}/2h}$$
以上不等式的指导意义是：
* 直升机的位置要合理，使$l$的值不至于过大；
* 应留有足够长的路程加速，使初速度尽可能大；

## 由方程(2)(3)(4)解得:

$$F_y >= (m\sqrt{2g(H+h)}-F_m\Delta t_2)/\Delta t_1$$

上式的指导意义是：
* 箱子的摆放应当尽可能增大竖直方向的作用力$F_y$
* 箱子应当有一定的承受能力，尽可能多地延长人/车与箱子的接触时间$\Delta t_1$
